```python
file_path = "/content/drive/MyDrive/Python/Global COVID-19 Dataset.csv"
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv(file_path, encoding='latin1')

print(df.head())
```

```python
df.rename(columns={
    'Country Name': 'Country',
    'Cases': 'Confirmed'
}, inplace=True)
df.drop(columns=['S. No.'], inplace=True)
print(df.head())
```

```python
top10 = df.sort_values(by='Confirmed', ascending=False).head(10)

print(top10)
```

```python
import matplotlib.pyplot as plt

top10.plot(x='Country', y='Confirmed', kind='bar')
plt.title("Top Countries by Cases")
plt.show()
```

```python
plt.pie(df['Confirmed'].head(5), labels=df['Country'].head(5), autopct='%1.1f%%')
plt.title("Top 5 Countries Distribution")
plt.show()
```

```python
plt.scatter(df['Confirmed'], df['Deaths'])
plt.xlabel("Confirmed")
plt.ylabel("Deaths")
plt.title("Confirmed vs Deaths")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
if 'Active' not in df.columns:
    df['Active'] = df['Confirmed'] - df['Deaths']
corr = df[['Confirmed', 'Deaths', 'Active']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
